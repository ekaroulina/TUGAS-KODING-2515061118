Matrices (Rank 2 Tensors) in NumPy

```python
import numpy as np
```

```python
# Membuat matrix X berukuran 2x3 menggunakan numpy
x = np.array([[1,10,8],[11,18,81]])
x
```

```python
# Menampilkan ukuran matriks x
x.shape
```

```python
#  Menampilkan seluruh elemen dari kolom pertama matriks x
x[:, 0]
```

```python
#  Menampilkan seluruh elemen dari baris kedua matriks x
x[1, :]
```

Matrices in PyTorch

```python
import torch
```

```python
# Membuat matrix X berukuran 2x3 menggunakan torch
x_pt = torch.tensor([[1,10,8],[11,18,81]])
x_pt
```

```python
# Menampilkan ukuran matriks x
x_pt.shape
```

```python
#  Menampilkan seluruh elemen dari kolom pertama matriks x dengan PyTorch
x_pt[:, 0]
```

```python
#  Menampilkan seluruh elemen dari baris kedua matriks x dengan Pytorch
x_pt[1, :]
```

Matrices in TensorFlow

```python
import tensorflow as tf
```

```python
# Membuat matriks X berukuran 2x3 dengan TensorFlow
x_tf = tf.Variable([[1,10,8],[11,18,81]])
x_tf
```

```python
# Menampilkan ukuran matriks x
tf.shape(x_tf)
```

```python
#  Menampilkan seluruh elemen dari kolom pertama matriks x dengan tensorflow
x_tf[:, 0]
```

```python
#  Menampilkan seluruh elemen dari baris kedua matriks x dengan tensorflow
x_tf[1, :]
```
